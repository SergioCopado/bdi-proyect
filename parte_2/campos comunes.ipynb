```python
import os
import json

# Ruta de la carpeta que contiene los archivos JSON
ruta_carpeta = "json"

# Lista para almacenar los campos comunes
campos_comunes = None

# Itera sobre todos los archivos en la carpeta
for nombre_archivo in os.listdir(ruta_carpeta):
    if nombre_archivo.endswith('.json'):
        ruta_completa = os.path.join(ruta_carpeta, nombre_archivo)
        
        # Abre el archivo JSON
        with open(ruta_completa, 'r') as f:
            data = json.load(f)
        
        # Obtiene los campos del primer elemento (si es una lista)
        campos_actuales = data[0].keys() if isinstance(data, list) else data.keys()
        
        # Si es el primer archivo, inicializa los campos comunes
        if campos_comunes is None:
            campos_comunes = set(campos_actuales)
        else:
            # Encuentra la intersección de los campos actuales con los campos comunes
            campos_comunes = campos_comunes.intersection(campos_actuales)

# Convierte la lista de campos comunes a una lista
campos_comunes = list(campos_comunes)

# Imprime los campos comunes
print("Campos comunes en todos los archivos:")
print(campos_comunes)
```

```python
import os
import json

def obtener_campos_json_en_carpeta(carpeta):
    campos = set()
    for archivo in os.listdir(carpeta):
        if archivo.endswith('.json'):
            ruta_archivo = os.path.join(carpeta, archivo)
            with open(ruta_archivo, 'r') as f:
                contenido = json.load(f)
                campos.update(obtener_campos_json(contenido))
    return campos

def obtener_campos_json(json_objeto):
    campos = set()
    if isinstance(json_objeto, dict):
        for key, value in json_objeto.items():
            campos.add(key)
            campos.update(obtener_campos_json(value))
    elif isinstance(json_objeto, list):
        for item in json_objeto:
            campos.update(obtener_campos_json(item))
    return campos

# Carpeta que contiene los archivos JSON
carpeta = "json"

# Obtener todos los campos presentes en los archivos JSON en la carpeta
campos = obtener_campos_json_en_carpeta(carpeta)

# Imprimir los campos encontrados
print("Campos encontrados en los archivos JSON:")
for campo in campos:
    print(campo)
```
